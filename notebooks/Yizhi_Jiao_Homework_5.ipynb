Problem 1

```python
def x(t, a, gamma, omega_0, phi):
    omega=(omega_0**2-gamma**2/4)**0.5
    return a*np.e**(-gamma*t/2)*np.cos(omega*t+phi)
time = np.linspace(1,1000,1000)/100
displacement_1 = x(time,1,1,1,-np.pi/2)
displacement_2 = x(time,1,0.5,1,-np.pi/2)
displacement_3 = x(time,1,1.5,1,-np.pi/2)
plt.title("Damped Oscillator with Different $\\gamma$ (A=1, $\\omega_0$=1, $\\phi$=1)")
plt.plot(time, displacement_1, color ="r", label="$\\gamma$ = 1")
plt.plot(time, displacement_2, color ="g", label="$\\gamma$ = 0.5")
plt.plot(time, displacement_3, color ="b", label="$\\gamma$ = 1.5")
leg = plt.legend(loc='upper center')
plt.ylabel('displacement / m')
plt.xlabel('time / s')
plt.show()
```

Problem 2(a)

Use Mathematica to find the first term of Taylor series $F_\nu$ at $x = 0$
$$F_\nu = \frac{2\pi k_B T \nu^2}{c^2}$$
so the power law of this relationship is $F_\nu \propto \nu^2$

```python
def F_nu(T, nu):
    k_B=1.380649e-23
    c=299792458
    return 2*np.pi*k_B*T*nu**2/c**2
```

Problem 2(b)

```python
nu = np.linspace(1e-9,1e9)
F_1 = F_nu(310.15, nu)
F_2 = F_nu(5778, nu)
plt.title("Blackbody Frequency Spectrum")
plt.plot(nu, F_1, color ="r", label="T = 310.15K")
plt.plot(nu, F_2, color ="g", label="T = 5778K")
plt.legend(loc='upper center')
plt.yscale('log')
plt.xscale('log')
plt.ylabel('$F_\\nu / Wm^{-2}Hz^{-1}$')
plt.xlabel('$\\nu / Hz$')
plt.show()
```

```python
from scipy.stats import linregress
x = np.log(nu)
y = np.log(F_1)
slope, intercept, r_value, p_value, std_err = linregress(x, y)
print(slope)
x = np.log(nu)
y = np.log(F_2)
slope, intercept, r_value, p_value, std_err = linregress(x, y)
print(slope)
```

The slope of the log vs log plot is 2.
This result agrees with my prediction.
