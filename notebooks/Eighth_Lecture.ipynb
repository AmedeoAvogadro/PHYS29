```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import interpolate
x = np.linspace(-4.4,4.4,10)
y = np.exp(-x**2)
x = np.linspace(-4.4,4.4,10)
plt.plot(x,y,linestyle=" ", marker='o')
y_interp = interpolate.interp1d(x,y)
xnew = np.linspace(-4.4,4.4,100)
ynew = y_interp(xnew)
y_lag = interpolate.lagrange(x,y)
y_cub = interpolate.interp1d(x,y, kind='cubic')
y_qua = interpolate.interp1d(x,y, kind='quadratic')
plt.plot(xnew,np.exp(-xnew**2))
plt.plot(xnew,ynew)
plt.plot(xnew,y_lag(xnew))
plt.plot(xnew,y_cub(xnew))
plt.plot(xnew,y_qua(xnew))
```

```python
x = np.linspace(-8, 8, 500)
y = np.linspace(-8, 8, 500)

x_grid, y_grid= np.meshgrid(x, y)
f = x_grid*np.exp(-0.5*np.sqrt(x_grid**2+y_grid**2))
plt.imshow(f,origin='lower',extent=[x[0],x[-1],y[0],y[-1]], cmap='magma')
plt.colorbar(label='$2p_x$')
plt.xlabel('x')
plt.ylabel('y')
```
