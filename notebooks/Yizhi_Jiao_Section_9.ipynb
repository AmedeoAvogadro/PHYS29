```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.integrate as integrate
plt.style.use('29S22.mplstyle')
```

```python
data=np.loadtxt('Sec9lengths.txt')
labels=data[:,0]
lengths=data[:,1]
```

```python
plt.plot(labels,lengths, linestyle=' ', marker='.')
plt.xlabel('labels')
plt.ylabel('lengths')
```

```python
plt.hist(lengths, bins=10)
```

```python
mean = np.mean(lengths)
sd = np.sqrt(np.var(lengths))

print(mean,'\n',sd)
```

```python
plt.errorbar(labels,lengths,yerr=sd, linestyle=' ', marker='o', label='measured')
```

```python
data=np.loadtxt('Sec9lengths_temps.txt')
time=data[:,0]
temp=data[:,1]
length=data[:,2]
```

```python
plt.plot(time,temp, color='c')
plt.errorbar(time, temp, yerr=1, linestyle=' ', marker='.')
```

```python
plt.plot(time, length, color='m')
plt.errorbar(time, length, yerr=5*10**-5, linestyle=' ', marker='.')
```

```python
plt.plot(temp, length, linestyle=' ', marker='o', color='y')
plt.errorbar(temp, length, xerr=1, yerr=5*10**-5, linestyle=' ', marker='.')
```

```python
newtemp = np.linspace(290,300,48)
alpha = (length-np.mean(length))/newtemp
plt.plot(newtemp, np.mean(length)+alpha*newtemp, linestyle=' ', marker='.')
```
