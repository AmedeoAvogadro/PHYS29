```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import special
x=np.linspace(0,10,6)
y=special.jv(0,x)
plt.plot(x,y,linestyle=' ',marker='o')
plt.xlabel('x')
plt.ylabel('$J_0(x)$')
plt.show() # Only if you need this command in your version of Jupyter
```

```python
from scipy import interpolate
xnew = np.linspace(0, 10, 100)
ynew = special.jv(0,xnew)
plt.plot(xnew, interpolate.interp1d(x, y)(xnew), color='r')
plt.plot(xnew, interpolate.interp1d(x, y, kind='cubic')(xnew), color='g')
plt.plot(xnew, interpolate.lagrange(x, y)(xnew), color='b')
plt.plot(xnew, ynew, color='k')
```

```python
x = np.linspace(-10, 10, 500)
y = np.linspace(-10, 10, 500)
x_grid, y_grid= np.meshgrid(x, y)
f = (2*(special.jv(1,np.sqrt(x_grid**2+y_grid**2)))/np.sqrt(x_grid**2+y_grid**2))
```

```python
plt.imshow(f,origin='lower',extent=[x[0],x[-1],y[0],y[-1]], cmap='plasma', vmax=0.1)
plt.colorbar()
plt.xlabel('x')
plt.ylabel('y')
```
