```python
def V(M,R,m,r):
    G=6.6743e-11 #Newton's gravitational constant in SI units
    if r<R:
        return -G*M*m*(3.*R**2-r**2)/2./R**3
    else:
        return -G*M*m/r
import matplotlib.pyplot as plt
import numpy as np
R=6.371e6 #earth radius in m
r=np.linspace(0.,3.e7) # create an array of 50 evenly spaced values between 0 and 3.e7
M=5.972e24 #earth mass in kg
m=1.
pot=0.*r
for i in range(50):
    pot[i]=V(M,R,m,r[i])
plt.plot(r,pot,color="y")
plt.xlabel('r')
plt.ylabel('V')
plt.axvline(x=R,linestyle='--',color="violet")
plt.show()
```

```python
x=np.linspace(-1,1)
P_0=x*0+1
P_1=x
P_2=0.5*(3*x**2-1)
P_3=0.5*(5*x**3-3*x)
plt.plot(x,P_0,color="c",linestyle="-.")
plt.plot(x,P_1,color="m",linestyle="--")
plt.plot(x,P_2, color="y",linestyle=":")
plt.plot(x,P_3, color="k")
plt.xlim(-1,1)
plt.xlabel('x')
plt.ylabel('P')
```

```python
def fibo(n):
    if n == 0:
        return 0
    elif n == 1:
        return 1
    else:
        return fibo(n-1)+fibo(n-2)
x=np.linspace(0,30,31)
y=np.linspace(0,0,31)
for i in range(0,31):
    y[i]=fibo(i)
plt.plot(x[2::],y[2::],marker="o",linestyle=" ")
plt.xlabel("n")
plt.ylabel("$x_n$")
plt.yscale('log')
```
