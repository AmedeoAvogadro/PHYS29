Problem 2

```python
x = np.linspace(-12, 12, 500)
y = np.linspace(-12, 12, 500)
x_grid, y_grid= np.meshgrid(x, y)
i_0 = 1
r = np.array([0.5, 1, 2])
i = i_0*np.sinc(x_grid/np.pi)**2*np.sinc(r[0]*y_grid/np.pi)**2
plt.imshow(i,vmax='0.005', extent=[x[0],x[-1],y[0],y[-1]])
plt.colorbar(label="Intensity")
plt.xlabel('x')
plt.ylabel('y')
```

```python
i = i_0*np.sinc(x_grid/np.pi)**2*np.sinc(r[1]*y_grid/np.pi)**2
plt.imshow(i,vmax='0.005', extent=[x[0],x[-1],y[0],y[-1]])
plt.colorbar(label="Intensity")
plt.xlabel('x')
plt.ylabel('y')
```

```python
i = i_0*np.sinc(x_grid/np.pi)**2*np.sinc(r[2]*y_grid/np.pi)**2
plt.imshow(i,vmax='0.005', extent=[x[0],x[-1],y[0],y[-1]])
plt.colorbar(label="Intensity")
plt.xlabel('x')
plt.ylabel('y')
```

For fixed horizontal width, the intensity increases then decreases as the vertical width increases.

Problem 3

```python
x = np.linspace(0, 1, 1000)
y = np.linspace(0, 1, 1000)
x_grid, y_grid = np.meshgrid(x, y)
t = 0
for n in range(1, 100, 2):
    t += 400 / (n*np.pi*np.sinh(n*np.pi)) * np.sinh(n*np.pi*y_grid) * np.sin(n*np.pi*x_grid)
plot = plt.contour(x_grid, y_grid, t, 11)
plt.clabel(plot, inline=1, fontsize=10)
```
